## Part 4 - Implement complete tic tac toe

```python
class ticTacToe():
    def __init__(self) -> None:
        self.emptySpaces = 0
        self.winCounters = [0,0]    #[p1,p2]
        self.win = False
        self.grid = []
        
    
    # create nxn empty grid default n=3
    def createGrid(self,n=3):
        grid = [[0 for _ in range(n)] for _ in range(n)]
        self.grid = grid
        self.emptySpaces = n**2

    # display grid default n=3
    def printGrid(self, n=3):
        """
        n type = int
        print nxn self.grid
        """
        # print cell
        top = " --- "  # top
        bottom = " --- "  # bottom

        dispGrid = [['' for _ in range(n)] for _ in range(n*3)]

        # create self.grid
        for i in range(1,n*3,3):
            for j in range(n):
                num = self.grid[(i-1)//3][j]
                # update cell with value num
                cell = f"| {num} |" 

                # save grid
                dispGrid[i-1][j] = top
                dispGrid[i][j] = cell
                dispGrid[i+1][j] = bottom
            
        # print self.grid
        for row in dispGrid:
            print(''.join(row))

    # check winner
    def checkWinner(self):
        n = len(self.grid)
        m = len(self.grid[0])
        if self.grid:
            # check horizontal
            for i in range(n):
                if all(x == self.grid[i][0] if x!= 0 else False for x in self.grid[i]):
                    # add win counter
                    self.winCounters[self.grid[i][0]] += 1
                    return f"Player {self.grid[i][0]} is the winner!"
            
            # check vertical
            for j in range(m):
                col = [self.grid[x][j] for x in range(n)]
                if all(x == col[0] if x!= 0  else False for x in col):
                    # add win counter
                    self.winCounters[self.grid[0][j]] += 1
                    return f"Player {self.grid[0][j]} is the winner!"
        
            # check diagonals (only two diagonals)
            if all(self.grid[i][i] == self.grid[0][0] if self.grid[i][i] != 0 else False for i in range(n)):
                # add win counter
                self.winCounters[self.grid[0][0]] += 1
                return f"Player {self.grid[0][0]} is the winner!"
            if all(self.grid[i][n-1-i] == self.grid[0][n-1] if self.grid[i][n-1-i] != 0 else False for i in range(n-1,0,-1)):
                # add win counter
                self.winCounters[self.grid[0][n-1]] += 1
                return f"Player {self.grid[0][n-1]} is the winner!"
        
        # If no winner yet, return false
        return False    
    
    # mark Cell from input
    def markCell(self,grid,player):
        """ 
        x int
        y int
        coordinates of the cell

        returns updated grid
        """
        # ask for place to put mark
        print("-- Input Help: The grid size is 3x3 --")
        try:
            x,y = tuple(input("Where do you want place your mark (row col): \nJust type the numbers"))
            
            # only mark place if its empty
            if grid[int(x)-1][int(y)-1] == 0:
                grid[int(x)-1][int(y)-1] = player
                self.emptySpaces -= 1
                return grid
            else:
                # run again
                print("Cell is already marked")
                return self.markCell(grid,player)
        except:
            print("Wrong input")
            return self.markCell(grid,player)

    # play game
    def play(self):
        players = [1,2]
        counter = -1
        cont = "Y"

        # create and show clean grid
        self.createGrid()
        self.printGrid()
        self.win = False
        
        # start game until empty or someone won
        while self.emptySpaces > 0 and self.win == False:
            # alternate player
            counter += 1
            player = players[(counter%2)]

            # player turn
            print(f"Player {player} turn:")
            self.grid = self.markCell(self.grid,player)
            # print grid
            self.printGrid()           

            # check if someone Won
            self.win = self.checkWinner()
        
        print(self.win if self.win else "Tie")

```

```python
# instance game and play
def main():
    game = ticTacToe()
    cont = "Y"
    while cont.upper() == "Y":
        game.play()
        print(f"Games won by Player 1: {game.winCounters[0]}")
        print(f"Games won by Player 2: {game.winCounters[1]}")

        cont = input("Want to play again? Y/N")


# run main
if __name__ == "__main__":
    main()
```
